### Real‑World Analogy: The Loan Officer

Imagine you're a **loan officer** at a bank. Your job: decide whether to approve a loan based on:

- **x₁**: Annual income ($)
    
- **x₂**: Credit score (0–1)
    
- **x₃**: Years at current job
    

Each applicant comes with these numbers. You need a **rule** to say YES (approve) or NO (deny).

**How does your brain decide?**

- If income is **high**, that's a strong signal → assign it **high importance** (weight)
    
- If credit score is **low**, that's a warning → assign it **negative importance** (negative weight)
    
- If years at job is medium, maybe not that important → assign **small weight**
    

You also have a **personal mood** (bias). If you're feeling generous today, you might approve borderline cases.

Your brain computes:

text

Score = (income × importance_income) + (credit × importance_credit) + (years × importance_years) + mood
If Score > 0 → Approve; else → Deny

**That's exactly a perceptron!** It's a simple decision maker that combines weighted evidence and adds a bias.

####  1.The `__init__` method – Setting up the learner
```python
import numpy as np
class Perceptron:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate # Step size for learning
        self.n_iterations = n_iterations   # how many times to see the data(
        self.weights = None                # The importance factors (to be learned)
        self.bias =None                    # The personal mood (to be learned)
```
**Real‑world analogy:**

- `learning_rate` = How quickly you change your mind. If it's high, one mistake makes you drastically change your opinion; if low, you're stubborn and change slowly.
    
- `n_iterations` = How many times you review the training examples. Like reading the same loan applications multiple times to really learn the pattern.**(How many times we go through the training data to improve the model)**
    
- `weights` = The importance factors for income, credit score, etc. Initially unknown.
    
- `bias` = Your starting mood before seeing any evidence.


#### 2.The `fit()` method - Learning from examples
`fit()` is a standard method name used in almost every ML model.
Its job is to train the model on data.

##### **What `fit()` does**

1. **Initializes the model**
    - Sets random weights for each feature
    - Sets bias to 0
2. **Looks at each example repeatedly**
    - For a fixed number of iterations (`n_iterations`)
3. **Makes predictions**
    - Computes weighted sum + bias → decides 0 or 1
4. **Checks mistakes**
    - Compares prediction to actual label
5. **Updates weights and bias**
    - Corrects model to be less wrong next time
6. **Repeats** until it learns patterns in the data

```python
def fit(self, X,y):
        """Train the perceptron on daya X and labels y.
        X: training samples(like loan applicants)
        Y: correct answeers (0=deny, 1=approve)
        """  
        n_samples,n_features = X.shape
        
        #initialize the weights to small random numbers
        self.weights = np.randn(n_features)* 0.01
        self.bias = 0
        
        #Loop over the dataset multiple times
        for _ in range(self.n_iterations):
            #look at each example one by one
            for idx, x_i in enumerate(X):
                #Compute the score (weighted sum +bias)
                linear_output = np.dot(x_i,self.weights) + self.bias
                
                #make a prediction (o or 1)
                y_predicted = self._step(linear_output)
           
                #if prediction is wrong, adjust weights and bias
                update = self.learning_rate * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update
```

**Real‑world analogy – Walking through this:**

Imagine you're training a new loan officer:

1. **Start with random opinions**: "I think income is super important, credit score not so much" (random weights).
    
2. **Look at first applicant**:
    
    - Actual outcome: They defaulted (y = 0, deny).
    - Your prediction: You approved (y_pred = 1) because your random weights said so.
    - **Mistake!** You need to adjust.
        
3. **Calculate the update**: `update = learning_rate × (correct - prediction)`
    - Here, correct=0, prediction=1 → (0 - 1) = -1
    - So update = 0.1 × (-1) = -0.1 (negative)
        
4. **Adjust weights**: For each feature, `weight += update × feature_value`
    
    - If income was high (x_i = 50,000), you subtract a lot → income becomes LESS important
    - If credit was low (x_i = 0.3), you subtract a little → credit becomes slightly LESS important
    - **Result**: You learn that high income + bad credit can still lead to default!
        
5. **Repeat** for every applicant, multiple times. Eventually, your weights settle on the correct pattern.
    

**key insight**: The update rule `update = learning_rate × (y_true - y_pred)` means:

- If prediction is correct (y_true - y_pred = 0) → no chang
- If prediction was 0 but should be 1 (under‑predicted) → positive update → increase weights
- If prediction was 1 but should be 0 (over‑predicted) → negative update → decrease weights


#### 3. The `predict()` Method – Making Decisions
```python
 def predict(self,X):             #  Predict class labels for samples in X.
        linear_output = np.dot(X, self.weights) + self.bias
        return self._step(linear_output)
```
**Real‑world analogy:**

- Take a new loan applicant, multiply their features by your learned weights, add bias.
    
- If the score > 0 → approve (1); else → deny (0).
    

It's like having a checklist where each factor is multiplied by its importance, and you sum them up.(**rm tis line later)

#### 4. The _step() Activation Function
```python
def _step(self, x):
    return np.where(x > 0, 1, 0)                #Step activation function: returns 1 if x > 0 else 0.
```
   ##### This is the decision rule:
- If the weighted sum + bias > 0 → output 1 (approve/positive class)
-  Otherwise → output 0 (deny/negative class)

##### Real‑world analogy:

The step function is like a light switch.
If the total score is positive → light ON (1, approve)
If the total score is negative or zero → light OFF (0, deny)

This is why a perceptron can only learn linear boundaries – it's like a single light switch that turns on when the weighted sum crosses zero.
